```python
#5.
data.fillna(data.mean(), inplace=True)
data.head(19)
```

```python
#6.Identify the outliers- Interquartile Range
Q1 = data['age'].quantile(0.25)
Q3 = data['age'].quantile(0.75)
IQR = Q3 - Q1
IQR
```

```python
sns.boxplot(data['age'],data=data)
```

```python
#6
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR
outliers = data[(data['age'] < lower_bound) | (data['age'] > upper_bound)]
outliers
```

```python
#6.Replace the outliers with a Median value
median_age = data['age'].median()
data.loc[(data['age'] < lower_bound) | (data['age'] > upper_bound), 'age'] = median_age
median_age
```

```python
#6.Verify the replacement
sns.boxplot(x=data['age'])
plt.show()
```

```python
#7.Check for categorical columns and perform encoding
categorical_cols = data.select_dtypes(include=['object']).columns
# Example: Perform one-hot encoding
encoded_data = pd.get_dummies(data, columns=categorical_cols)
encoded_data
```

```python
#8. Split the data into dependent and independent variables
X = encoded_data.drop('survived', axis=1)
y = encoded_data['survived']
X
```

```python
y
```

```python
#9.Scale the independent variables
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
X_scaled
```

```python
#10. Split the data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)
```
